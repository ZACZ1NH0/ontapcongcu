## Nhap ma tran

```python
import numpy as np

A = np.array([
    [1,2,3,2],
    [2,3,5,7],
    [4,5,2,1],
    [9,8,2,1]
])
```

```python
A
```

```python
from scipy.fftpack import fft, dct, idct, ifft

A_dct = dct(A, type = 2, norm='ortho')
```

```python
A_dct
```

```python
A_idct = idct(A_dct, type = 2, norm='ortho')
A_idct
```

```python
A_fft = np.fft.fft2(A, norm = 'ortho')
```

```python
A_fft
```

```python
A_ifft = np.fft.ifft2(A_fft, norm = 'ortho')
```

```python
A_ifft
```

```python
A_10 = A+10
```

```python
A_10
```

```python
print(A**3)
```

```python
C = np.sum(A<5)
```

```python
C
```

```python
import scipy.linalg as la

results = la.eig(A)
print('vector rieng',results[1])
print('gia tri rieng: ',results[0])
```

#### tinh sum, mean, std, max, min theo hang, theo cot, toan ma tran

Toàn bảng

```python
print('sum: ', np.sum(A))
print('mean: ', np.mean(A))
print('max: ', np.max(A))
print('min: ', np.min(A))
```

Theo cột

```python
print('sum: ', np.sum(A, axis = 0))
print('mean: ', np.mean(A, axis = 0))
print('max: ', np.max(A, axis = 0))
print('min: ', np.min(A, axis = 0))
```

Theo hàng


```python
print('sum: ', np.sum(A, axis = 1))
print('mean: ', np.mean(A, axis = 1))
print('max: ', np.max(A, axis = 1))
print('min: ', np.min(A, axis = 1))
```

## Tinh tich phan

```python
from scipy import integrate
f = lambda x: x**2 + np.sin(x)
result = integrate.quad(f, 0, np.pi)
print(f"Ket qua tich phan: {result}")
```

### Tính đạo hàm

```python
import sympy
x = sympy.Symbol('x')
y = x**4 - 5*x**2 +20*sympy.cos(x)
y1 = y.diff(x)
print(y1)
for i in range (-10, 10):
    print(y1.subs(x,i).evalf())
```

### Tính lim

```python
import sympy
x = sympy.symbols('x')
y = (3*x**2 + 2*x -5)/(x**3+7*x**2+1)
result = sympy.limit(y, x, np.inf)
result
```

### Gia tri lon nhat, be nhat cua ham so tren doan

```python
from scipy.optimize import fminbound
import math

def f(x): return np.cos(x)**2 + x**3 - 5*x
min_loc = fminbound(f, -2, 3)
print(min_loc)
```

## Giai he phương trình

```python
a = np.array([[1,2,1],[1,3,2],[1,1,-1]], dtype=float)
b = np.array([[6],[3],[4]], dtype=float)
sol = np.linalg.solve(a,b)
x, y, z = sol
print(f"Nghiem cua he phuong trinh {x} ,{y} ,{z}")
```
