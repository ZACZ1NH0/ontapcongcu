```python
import math
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model(y, x):
    dydx = x**2
    return dydx
y0 = 5
x = np.linspace(0, 100)
y = odeint (model, y0, x)
plt.plot(x,y)
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
def model(y, x):
    dydx = -x/(1-x**2) * y + x*math.sqrt(y)
    return dydx

y0 = 2
x = np.linspace(4,10)
y = odeint(model, y0, x)
plt.plot(x,y)
plt.xlabel('Truc x')
plt.ylabel('Truc y')
plt.show()
```

```python
def model (U, x):
    return [U[1], (1 - U[0]**2)*U[1] - U[0]]

U0 = [2, 0]
xs = np.linspace(5, 100)
Us = odeint(model, U0, xs)
ys = Us[:,0]
plt.xlabel("truc x")
plt.ylabel("truc y")
plt.title("PTVP")
plt.plot(xs, ys)
plt.show()
```

```python
#Bai 2:  tich phan 0 ->pi f = (x^2 +sin(x))
import math 
from scipy import integrate
f = lambda x: x**2 + math.sin(x)
result = integrate.quad(f, 0, math.pi)
print(f"Ket qua tich phan: {result}")
```

```python
#Bai 4: Tinh gioi han
import math
import sympy
import matplotlib.pyplot as plt
x = sympy.symbols('x')
y = (3*x**2 + 2*x -5)/(x**3+7*x**2+1)
result = sympy.limit(y, x, math.inf)
print(f'Gioi han cua ham y tu x den vo cung: {result}')

x = np.arange(-20,20,0.1)
y = (3*x**2 + 2*x -5)/(x**3+7*x**2+1)
plt.plot(x,y)
plt.xlabel("truc x")
plt.ylabel("truc y")
plt.show()
```

```python
# Bai 5: Gia tri lon nhat cua ham so y = np.cos(x)**2 + x**3 -5*x  tren doan -2 den 3

import math
from scipy.optimize import fminbound

def f(x): return (math.cos(x))**2 + x**3 - 5*x
min_loc = fminbound(f, -2, 3)
print(min_loc)
```
